<h4>Import bibliotek</h4>

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set_style("darkgrid")
```

<h4>Wczytanie plików do analizy</h4>

```python
qual_df1 = pd.read_excel("Metabolomika_Quality1.xlsx")
qual_df2 = pd.read_excel("Metabolomika_Quality2.xlsx", sheet_name="Arkusz2")
```

<h4>Analiza PCA</h4>

```python
qual_df1
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# First, let's prepare the data for PCA
# We need to exclude non-numeric columns like ID and Group
X = qual_df1.drop(['Id', 'Group'], axis=1)

# Standardize the data (important for PCA)
X_scaled = StandardScaler().fit_transform(X)

# Perform PCA
pca = PCA(n_components=10)  # Let's start with 2 components
principal_components = pca.fit_transform(X_scaled)

# Create a DataFrame with the principal components
pca_df1 = pd.DataFrame(data=principal_components, columns=[f"PC{i}" for i in range(len(pca.components_))])
pca_df1['Id'] = qual_df1['Id']
pca_df1['Group'] = qual_df1['Group']

# Let's see how much variance is explained by each component
explained_variance = pca.explained_variance_ratio_
print(f"Explained variance by PC1: {explained_variance[0]:.2%}")
print(f"Explained variance by PC2: {explained_variance[1]:.2%}")
print(f"Total explained variance: {sum(explained_variance):.2%}")

# Plot the PCA results
plt.figure(figsize=(10, 8))
groups = pca_df1['Group'].unique()
colors = sns.color_palette('husl', n_colors=len(groups))

for i, group in enumerate(groups):
    group_data = pca_df1[pca_df1['Group'] == group]
    plt.scatter(group_data['PC1'], group_data['PC2'], 
                label=group, color=colors[i], s=60, alpha=0.7)

plt.title('PCA of Metabolomics Data', fontsize=15)
plt.xlabel(f'PC1 ({explained_variance[0]:.2%})', fontsize=12)
plt.ylabel(f'PC2 ({explained_variance[1]:.2%})', fontsize=12)
plt.legend(title='Group')
plt.grid(True, linestyle='--', alpha=0.7)

# Add a tight layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Let's create a more detailed visualization of PCA results

# 1. Create a scree plot to visualize the explained variance
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
explained_variance = pca.explained_variance_ratio_
plt.bar(range(1, len(explained_variance) + 1), explained_variance, alpha=0.7)
plt.step(range(1, len(explained_variance) + 1), np.cumsum(explained_variance), where='mid', color='red', linestyle='--')
plt.ylabel('Explained Variance Ratio')
plt.xlabel('Principal Component')
plt.title('Scree Plot')
plt.grid(True, linestyle='--', alpha=0.5)
plt.xticks(range(1, len(explained_variance) + 1))

# 2. Improve the scatter plot with more details
plt.subplot(1, 2, 2)
groups = pca_df1['Group'].unique()
colors = sns.color_palette('viridis', n_colors=len(groups))
markers = ['o', 's', '^', 'D', 'v', '<', '>', 'p', '*']

for i, group in enumerate(groups):
    group_data = pca_df1[pca_df1['Group'] == group]
    plt.scatter(group_data['PC1'], group_data['PC2'], 
                label=group, color=colors[i], 
                s=80, alpha=0.8, marker=markers[i % len(markers)],
                edgecolor='w', linewidth=0.5)

# Add sample IDs as annotations
for i, txt in enumerate(pca_df1['Id']):
    plt.annotate(txt, (pca_df1['PC1'].iloc[i], pca_df1['PC2'].iloc[i]),
                 fontsize=8, alpha=0.7)

plt.title('PCA of Metabolomics Data', fontsize=14)
plt.xlabel(f'PC1 ({explained_variance[0]:.2%})', fontsize=12)
plt.ylabel(f'PC2 ({explained_variance[1]:.2%})', fontsize=12)
plt.legend(title='Group')
plt.grid(True, linestyle='--', alpha=0.5)
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)

plt.tight_layout()
plt.show()

# Print the top 10 components' explained variance
print("Explained variance by top 10 components:")
for i, var in enumerate(explained_variance[:10], 1):
    print(f"PC{i}: {var:.2%} (cumulative: {sum(explained_variance[:i]):.2%})")
```
