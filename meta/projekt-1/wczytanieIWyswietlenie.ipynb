#### Importowanie bibliotek

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("darkgrid")
```

#### Wczytanie zbioru danych

```python
df = pd.read_csv("Metabolomics_Final.csv", sep=";")
```

#### Usunięcie NaN i Null

```python
df = df.dropna(axis=1)
```

```python
df.shape
```

#### Zapisanie zbioru danych bez NaN

```python
df.to_csv("Metabolomics_Final_clean.csv", index=False)
```

#### Statystyki dla obiektów

```python
df.describe()
```

#### Standaryzacja zbioru

```python
from sklearn.preprocessing import StandardScaler

# Select only numeric columns for standardization
numeric_columns = df.select_dtypes(include=['int64', 'float64']).columns

# Create and fit the scaler
scaler = StandardScaler()

# Transform the numeric columns
df[numeric_columns] = scaler.fit_transform(df[numeric_columns])
```

```python
df.head()
```

```python
df.describe()
```

#### Wyświetlenie rozkładów zmiennych dla grup

```python
# Melt the dataframe to long format for easier plotting
# Exclude non-metabolite columns (Name and Group)
melted_df = pd.melt(df, id_vars=['Name', 'Group'], value_vars=numeric_columns,
                    var_name='Metabolite', value_name='Value')

# Create a figure with appropriate size
plt.figure(figsize=(14, 8))

# Create box plot by Group
sns.boxplot(x='Group', y='Value', data=melted_df)

# Add a title and labels
plt.title('Distribution of Metabolite Values by Group', fontsize=16)
plt.xlabel('Group', fontsize=14)
plt.ylabel('Standardized Value', fontsize=14)
plt.xticks(rotation=45)

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```
