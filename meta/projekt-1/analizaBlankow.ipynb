#### Importowanie bibliotek

```python
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt

sns.set_style("darkgrid")
```

#### Wczytanie zbioru danych

```python
df = pd.read_csv("Metabolomics_Final_clean_filtered.csv")
```

```python
df.head()
```

#### Filtrowanie blanków ze zbioru danych

```python
df = df[df['Group'] == 'Blank']
```

```python
df.head()
```

```python
df.shape
```

#### Obliczenie macierzy korelacji dla obiektów

```python
# Drop non-numeric columns (Name and Group) and transpose the DataFrame
numeric_df = df.drop(['Name', 'Group'], axis=1)
correlation_matrix = numeric_df.T.corr()

# Optionally, display the correlation matrix
print("Correlation matrix shape:", correlation_matrix.shape)
correlation_matrix
```

```python
correlation_matrix.describe()
```

#### Heatmapa macierzy korelacji

```python
# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, 
            cmap='coolwarm', 
            center=0,
            annot=True,
            fmt='.2f',
            square=True)

plt.title('Correlation Matrix Heatmap')
plt.tight_layout()
plt.show()
```

Wysokie wartości korelacji między blankami wskazuje na poprawne przygotowanie aparatury do pomiarów
